```python
data.dropna(subset=['Arrival Delay in Minutes'],inplace=True)
```

```python
# Label Encoding of object datatype
for col in data.select_dtypes(include=['object']).columns:
  print(f"{col}: {data[col].unique()}")
```

```python
# import LabelEncoder
from sklearn.preprocessing import LabelEncoder
for col in data.select_dtypes(include=['object']).columns:
  # loading the LabelEncoder object
  label_encoder = LabelEncoder()

  # fitting the unique values
  label_encoder.fit(data[col].unique())

  # transform the fitted values
  data[col] = label_encoder.transform(data[col])

  print(f"{col}: {data[col].unique()}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(),annot=True,fmt='.2g')
```

```python
data.columns
```

```python
# Train Test Split
X = data.drop('satisfaction',axis=1)
y = data['satisfaction']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, precision_score, f1_score, recall_score, log_loss
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# removing the outliers
from scipy import stats

# selecting the columns
selected_columns = ['Departure Delay in Minutes','Arrival Delay in Minutes']

# calculating the Z Score
z_score = np.abs(stats.zscore(X_train[selected_columns]))

# setting up threshold value
threshold_val = 3

# find the indices of outliers based on the threshold value
outlier_indices = np.where(z_score > threshold_val)[0]

# removing the outliers
X_train = X_train.drop(X_train.index[outlier_indices])
y_train = y_train.drop(y_train.index[outlier_indices])
```

```python
# Model Building

# 1. Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

dtree = DecisionTreeClassifier(class_weight='balanced')

# setting up the parameters
param_grid = {
    'max_depth':[3,4,5,6,7,8],
    'min_samples_split':[2,3,4],
    'min_samples_leaf':[1,2,3,4],
    'random_state':[0,42]
}

grid_search = GridSearchCV(dtree, param_grid, cv=5)
grid_search.fit(X_train, y_train)

print(grid_search.best_params_)
```

```python
dtree = DecisionTreeClassifier(max_depth= 8, min_samples_leaf= 2, min_samples_split= 2, random_state= 42, class_weight='balanced')
dtree.fit(X_train, y_train)
```

```python
dtree_pred = dtree.predict(X_test)
print("Acuracy Score: ",round(accuracy_score(y_test, dtree_pred)*100, 2), "%")
```

```python
from sklearn.metrics import accuracy_score, classification_report, precision_score, f1_score, recall_score, log_loss
```

```python
print("Classification Report", classification_report(y_test, dtree_pred))
print("Precision Score", precision_score(y_test, dtree_pred))
print("F1 Score", f1_score(y_test, dtree_pred))
print("Recall Score", recall_score(y_test, dtree_pred))
print("Log Loss", log_loss(y_test, dtree_pred))
```

```python
imp_data = pd.DataFrame({
    "Feature Name": X_train.columns,
    "Importance":dtree.feature_importances_
})

fi = imp_data.sort_values(by="Importance", ascending=False)
fi2 = fi.head(10)
plt.figure(figsize=(10,8))
sns.barplot(data=fi2, x="Importance",y="Feature Name")
plt.title("Top 10 Fetaures Names")
plt.xlabel("Importance",fontsize=16)
plt.ylabel("Feature Name", fontsize=16)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, dtree_pred)
plt.figure(figsize=(10,8))
sns.heatmap(data=cm, linewidths=.5, annot=True, cmap="Blues")
plt.xlabel("Actual Label",fontsize=16)
plt.ylabel("Predicted Label", fontsize=16)
smapel_title = "Accuracy Score for Decision Tree: {0}".format(dtree.score(X_test, y_test))
plt.title(smapel_title, size=15)
```

```python
# 2. Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rfc = RandomForestClassifier(class_weight='balanced')

# setting up the parameters
param_grid = {
    'n_estimators':[100,200],
    'max_depth':[None, 5, 10],
    'max_features':['sqrt', 'log2', None],
    'random_state':[0,42]
}

grid_search = GridSearchCV(rfc, param_grid, cv=5)
grid_search.fit(X_train, y_train)

print(grid_search.best_params_)
```

```python
rfc = RandomForestClassifier(max_depth= 8, min_samples_leaf= 2, min_samples_split= 2, random_state= 42, class_weight='balanced')
rfc.fit(X_train, y_train)
```

```python
ref_pred = rfc.predict(X_test)
print("Acuracy Score: ",round(accuracy_score(y_test, ref_pred)*100, 2), "%")
```

```python
from sklearn.metrics import accuracy_score, classification_report, precision_score, f1_score, recall_score, log_loss
print("Classification Report", classification_report(y_test, ref_pred))
print("Precision Score", precision_score(y_test, ref_pred))
print("F1 Score", f1_score(y_test, ref_pred))
print("Recall Score", recall_score(y_test, ref_pred))
print("Log Loss", log_loss(y_test, ref_pred))
```

```python
imp_data = pd.DataFrame({
    "Feature Name": X_train.columns,
    "Importance":rfc.feature_importances_
})

fi = imp_data.sort_values(by="Importance", ascending=False)
fi2 = fi.head(10)
plt.figure(figsize=(10,8))
sns.barplot(data=fi2, x="Importance",y="Feature Name")
plt.title("Top 10 Fetaures Names")
plt.xlabel("Importance",fontsize=16)
plt.ylabel("Feature Name", fontsize=16)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, ref_pred)
plt.figure(figsize=(10,8))
sns.heatmap(data=cm, linewidths=.5, annot=True, cmap="Blues")
plt.xlabel("Actual Label",fontsize=16)
plt.ylabel("Predicted Label", fontsize=16)
smapel_title = "Accuracy Score for Random Forest Classifier: {0}".format(dtree.score(X_test, y_test))
plt.title(smapel_title, size=15)
```
